```python
import seaborn as  sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
import random
x= ['a','b','c','b','c','a','a']
y= [random.randint(20,80)for i in range (7)]
z= ['m','s','s','h','m','h','en']
df = pd.DataFrame({"Name":x,"Marks":y,"subject":z})
print(df)
plt.figure(figsize=(8,3))
sns.barplot(x='Name',y="Marks",data=df,hue="subject")
plt.legend(loc="upper right")
plt.plot()
```

```python
plt.figure(figsize=(9,3))

sns.set_style("darkgrid")
sns.lineplot(x='tip',y='total_bill',hue='day',data=df)
```
