```python
import random
maths= [random.randint(0,101)for i in range(5)]
eng = [random.randint(0,101)for i in range(5)]
hindi= [random.randint(0,101)for i in range(5)]
names=["a","b","c","d","e"]
width=0.3
plt.figure(figsize=(5,5))
MathIdx = [i for i in range(5)]
EngIdx= [mathIdx + width for mathIdx in range(5)]
HindiIdx= [EngIdx + 2*width for EngIdx in range(5)]

plt.bar(MathIdx,maths,color="r",width=width)
plt.bar(EngIdx,eng,color="b",width=width)
plt.bar(HindiIdx,hindi,color="k",width=width)
plt.xticks(MathIdx,names)
plt.legend(["Math","English","Hindi"],loc="upper right")
plt.show()
           
           
```

```python
import random
maths= [random.randint(0,101)for i in range(5)]
eng = [random.randint(0,101)for i in range(5)]
hindi= [random.randint(0,101)for i in range(5)]
names=["a","b","c","d","e"]
width=0.3
plt.figure(figsize=(5,5))
MathIdx = [i for i in range(5)]
EngIdx= [mathIdx + width for mathIdx in range(5)]
HindiIdx= [EngIdx + 2*width for EngIdx in range(5)]

plt.bar(MathIdx,maths,color="r",width=width)
plt.bar(EngIdx,eng,color="b",width=width)
plt.bar(HindiIdx,hindi,color="k",width=width)
plt.xticks(MathIdx,names)
plt.legend(["Math","English","Hindi"],loc="upper right")
plt.savefig("mygraph")
plt.show()
           
```

  ##PIE CHART

```python
import random
x=[random.randint(30,101) for i in range(5)]
print(x)
y=['math','eng','chem','hindi','sans']
plt.pie(x,labels=y,shadow='gray',colors=['g','r','y','b','k'])
plt.show()
```

```python
#### explode function
import random
x=[random.randint(30,101) for i in range(5)]
print(x)
y=['math','eng','chem','hindi','sans']
ex=[0.2,0,0,0,0]
plt.pie(x,labels=y,shadow='gray',colors=['g','r','y','b','k'],explode=ex)
plt.show()
```

```python
##starting agle
import random
x=[random.randint(30,101) for i in range(5)]
print(x)
y=['math','eng','chem','hindi','sans']
ex=[0.2,0,0,0,0]
plt.pie(x,labels=y,shadow='gray',colors=['g','r','y','b','k'],explode=ex,startangle=0)
plt.title("my pie chart")
plt.show()
```
