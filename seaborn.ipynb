```python
###SEABORN
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#helb(sns)
```

```python
df= sns.load_dataset("tips")
print(df)
```

```python
sns.stripplot(x='tip',y='sex',data=df)
plt.plot()
```

```python
##STRIPPLOT
x=[1,2,3,4,5]
y=[9,0,4,7,2]
ds =pd.DataFrame({"a":x,"b":y})
sns.stripplot(ds)
```

```python
##SCATTERPLOT
plt.figure(figsize=(5,2))
sns.scatterplot(x='tip',y='total_bill',data=df,hue='day',size=0.2,style='day',palette='Accent',legend='brief')
```

```python
##LINEPLOT
plt.figure(figsize=(6,3))
color = ['green','Blue','Red','Yellow']
sns.set_palette(color)
sns.set_style("darkgrid")
sns.lineplot(x='tip',y='total_bill',data=df,marker='D',hue='day')           
```
